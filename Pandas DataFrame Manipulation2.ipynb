```python
# By default index gets populated
ser1 = pd.Series(np.array([4,5,6,7,8]))
ser1
```

```python
#We can give our own index while matching the data. It is called Named indexes

ser1 = pd.Series(np.array([4,5,6,7,8]),index=['a','b','ravi','kumar','Dubey'])
ser1
```

```python
#we can access data at index 'Ravi'

ser1['ravi']
```

```python
# We cannot acess the series with named index using ILOC
# ser1.iloc["ravi"]  --- It will throw error as iloc can only retrieve using numeric index value which gets assigned by default

ser1.iloc[2]
```

```python
# We can give named index while accessing the data using LOC function

ser1.loc['ravi']

#ser1.loc[2]  #it will throw error 
```

```python
#lets create another series and give named indexes as numeric value
ser2 = pd.Series(np.array([4,5,6,7,8]),index=[45,52,67,78,99])
ser2
```

```python
# We can access value at 4th Index using named index 78

ser2[78]
```

```python
# ILOC will not work with named index even if it is integer. Instead it will work only with actual/default index 3

ser2.iloc[3]
```

```python
#LOC will work with named index
ser2.loc[45]
```

```python
#we can give user defined indexes and column name

import random
df2 = pd.DataFrame(np.random.randint(13,18,(4,5)),index=['a1','b1','c1','d1'],columns=['v','w','x','y','z'])
df2
```

```python
# in case we need particular values from dataset df2 we can choose loc function

df2.loc[["c1","d1"],["x","y"]]
```

```python
# in case we need particular values from dataset df2 we can choose iloc function but have to give actual indexes

df2.iloc[[2,3],[2,3]]
```

```python
#we can acces using below options 

df2.iloc[2:,2:4]
```

```python
#in case we need summation of column x and z and stores values in new column name new_col_x+z
df2["new_col_x+z"] = df2["x"] + df2["z"]
df2
```

```python
# by default axis = 0  which is meant for column
# using default setup we can drop any row value

#to drop a column we have to change axis from 0 to 1 to refer to column 

df2.drop("new column",axis = 1)
```
