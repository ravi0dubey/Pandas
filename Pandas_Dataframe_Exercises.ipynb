```python
#Reading CSV from github repository in raw format and it directly convert into dataframe
#Read CSV file which is separated by Tab

import pandas as pd
data1= pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep ='\t')
```

```python
#converting data into CSV and storing in local system for future references
data1.to_csv("chipotle.csv")
```

```python
data1.dtypes
```

```python
#1. Find the Maximum price of itemName
data1['item_price'].str.startswith('$')
```

```python
# Currently price is string so creating a new list  which stores Numeric value of price
list1= []
for i in data1['item_price']:
    list1.append(float(i[1:]))

data1["Price_in_Numeric"] = list1
data1
```

```python
list1
```

```python
#1. Find the Maximum price of itemName
max(data1['Price_in_Numeric'])
```

```python
#2. Find the item_name which has maximum price  without displaying the Price_in_Numeric column
data1.loc[data1['Price_in_Numeric'] == max(data1['Price_in_Numeric']),['order_id','quantity', 'item_name','choice_description','item_price']]
```

```python
#3. Get the list of Item_name where choice_description is not blank
data1[data1["choice_description"].isnull()==False]
```

```python
data1["choice_description"]
```

```python
# 4 Getting length of Description item

# for i in data1[data1["choice_description"].isnull() == True]['choice_description']:
#     print(f"{i},{type(i)},{str(i)},{type(str(i))}")
list2= []
for i in data1["choice_description"]:
    print(f"Each data frame : {i}, {len(str(i))}")
    if str(i) =="nan":
        list2.append(0)
    else:
        list2.append(len(str(i)))
```

```python
# 4 inserting column in between the data1
data1.insert (4, "Choice_Description_Length", list2)
data1
```

```python
# 5 Get unique item_name
list_item_name = []
for i in data1["item_name"].unique():
    list_item_name.append(i)
list_item_name
```

```python
# 6 Generate list of items where order_quantity is greater than 1
data1[data1["quantity"] > 1]
```

```python
# 7 generate list of items where order_quantity is greater than 1 and Item price is less than $20
data1[(data1["Price_in_Numeric"] >10.00) & (data1["quantity"] > 1)]
```

```python
# 7 Generate list of items where order_quantity is greater than 1 and Item price is less than $20
#in case we just want intial columns of data then we can either use loc function 

data1.loc[(data1["Price_in_Numeric"] >10.00) & (data1["quantity"] > 1),['order_id','quantity', 'item_name','choice_description','item_price']]
```

```python
#7 Generate list of items where order_quantity is greater than 1 and Item price is > $10
#in case we just want intial columns of data then we can give column name directly

data1[(data1["Price_in_Numeric"] >10.00) & (data1["quantity"] > 1)][['order_id','quantity', 'item_name','choice_description','item_price']]
```

```python
# 8 Generate list of items where order_quantity is greater than 1 and Item price is > $20 and Starts with 'Fresh'
#in case we just want intial columns of data then we can give column name directly
data1[(data1["Price_in_Numeric"] >10.00) & (data1["quantity"] > 1) & (data1['choice_description'].str.startswith('[Fresh'))] [['order_id','quantity', 'item_name','choice_description','item_price']]
```

```python
#9 Generate list of items where order_quantity is greater than 1 and Item price is less than $20 and has Rice in it
#in case we just want intial columns of data then we can give column name directly
data1[(data1["Price_in_Numeric"] < 20.00) & (data1["quantity"] > 1) & (data1["choice_description"].isnull()==False) & data1['choice_description'].str.contains('Rice')] [['order_id','quantity', 'item_name','choice_description','item_price']]
```

```python
#10 Get list of records with Item name  Having Chicken in it, Price > 10 and Quantity > 1
data1[(data1['item_name'].str.contains('Chicken') == True) & (data1["Price_in_Numeric"] > 10.00) & (data1["quantity"] > 1)]
```
