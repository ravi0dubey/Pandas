# SERDE stands for Serialization and Deseralization

where Serialization is conversion of object to bytecode

Deserialization is conversion of bytecode to Object

Pickle and HD5 follows SERDE


```python
import pandas as pd
```

```python
#It will read a csv file and store data in serialized format using to_pickel.
# pickle1 stores data in bytecode

data1 = pd.read_csv("D:\\Study\\Data Science\\Python\\ineuron\\html0.csv")
data1.to_pickle("pickle1")
```

```python
data1
```

# HD5 or .H5 File Format

```python
#It will create af file test in h5 format

data2 = pd.HDFStore('test.h5')
```

```python
data2
```

```python
data2['column1']= data1
```

```python
#dataset is stored in serialized form
#it works in key and value pair

data2['column1']
```

```python
#data1 records whcih was read from csv is put in data2 in tabular format

data2.put('obj2',data1,format = 'table')
```

```python
#retrieving records from data2

data2.select('obj2', where = ['index > 2 and index <= 7 '])
```
