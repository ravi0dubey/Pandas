```python
#create an inbuilt function and apply it to dataframe values
def profit_flag(a):
    if a>=0:
        return 'Positive Profit'
    else:
        return 'Negative Profit'

profit_flag(-150)
```

```python
profit_flag(df16['profit'][1])
```

```python
# using profit values call function profit_flag 

df16['Flag_Profit']= df16['profit'].apply(profit_flag)
```

```python
df16
```

```python
df16.groupby('Flag_Profit')['profit'].count()
```

```python
#we can call inbuilt function len to fetch length of customer name

df16['Len_customer_name']= df16['customer_name'].apply(len)
df16
```

```python
def quantity_flag(qty):
    if qty <= 5:
        return 'Low'
    elif qty>5 and qty < 8:
        return 'Medium'
    else:
        return 'High'

quantity_flag(1)
```

```python
df16['flag_Qty']= df16['quantity'].apply(quantity_flag)
df16
```

```python
df16['Square_Quantity']= df16['quantity'].apply(lambda a: a**2)
df16
```
