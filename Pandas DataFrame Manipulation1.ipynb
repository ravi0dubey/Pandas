```python
#this is our dataset with values in it

pd.Categorical(["Ravi","Astha","Illisha","Lika"])
```

```python
# default property of dataset dataset with ordered = False
pd.Categorical(["Ravi","Astha","Illisha","Lika"],ordered=False)
```

```python
# When we give Ordered = True, although it sorts data in alphabetical order and gives ranking to the data
pd.Categorical(["Ravi","Astha","Illisha","Lika"],ordered=True)
```

```python
# Apart from Ordered = True, When we insert a new condition with categories, this specify the the ranking within the data 

pd.Categorical(["Ravi","Astha","Illisha","Lika"],categories=["Lika","Illisha","Astha","Ravi"],ordered=True)
```

```python
#With Ordered = False, with new condition categories, it sorts data in the order mentioned in categories 
pd.Categorical(["Ravi","Astha","Illisha","Lika"],categories=["Lika","Illisha","Astha","Ravi"],ordered=False)
```
