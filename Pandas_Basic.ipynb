# Saving Data in CSV Format and Reading the same CSV


```python
#data4.csv will be created without Header
dt7.to_csv("data4.csv",header= False)
```

```python
#data5.csv will be created only with column empid and empsal
# If index = False, first column which gets generated by default for index will not be created
dt7.to_csv("data5.csv",columns=['empid','empsal'], index = False)
```

```python
#data6 will be created with separator '?' instead of ','
dt7.to_csv("data6.csv",sep= '?',  index = False)
```
