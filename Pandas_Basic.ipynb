# 3 approach of Reading 2nd Nested Level record

```python
#APPROACH 1

#Reading it without Pandas
#User column is at 2nd nested level and in case we need to read details of all users

for i in range(len(s3.json())):
    print(s3.json()[i]['user'])
```

```python
#Reading it without Pandas
#User column is at 2nd nested level and in case we need to read login id of all users

for i in range(len(s3.json())):
    print(s3.json()[i]['user']['login'])
```

```python
#Approach 2

#Using Pandas
#User column is at 2nd nested level and in case we need to read login id of all users

l = []
for i in range (len(s3.json())):
    l.append((s3.json()[i]['user']))

#it will display all fields
pd.DataFrame(l)
```

```python
#Using Pandas
#User column is at 2nd nested level and in case we need to read login id of all users

l = []
for i in range (len(s3.json())):
    l.append((s3.json()[i]['user']))


#it will display columns Login
pd.DataFrame(l, columns=['login'])
```

```python
#Approach 3

#using Pandas without storing into list

#Reading api link for JSON data without using API and it can convert directly into dataframe for 2nd nested level of user

import json
s5= pd.read_json('https://api.github.com/repos/pandas-dev/pandas/issues')
s6= pd.DataFrame.from_records(s5['user'])
s6.head()
```
